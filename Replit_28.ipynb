```python
from IPython.display import clear_output
import time, random

print("⚔️ Character Builder")

##function to roll dice
def rollDice(side):
  result = random.randint(1,side)
  return result

##function to get health stat
def healthCalc():

    healthst = (rollDice(6) + rollDice(12))/2 + 10

    return healthst

##function to get strength stat
def strengthCalc():

    strengthst = (rollDice(6) + rollDice(12))/2 + 12

    return strengthst

def damageCalc(winnerSt, loserSt):
    damage = abs(winnerSt - loserSt) + 1
    return damage 

##while loop to generate character

i = 0
while i < 1:
    clear_output(wait=True)
    legendName1 = input("What is the legend's name?")
    legendType1 = input("What is the legend's character type?")

    strength1 = strengthCalc()
    health1 = healthCalc()

    print("Character Name:",legendName1)
    time.sleep(1)
    print("Strength:",strength1)
    time.sleep(1)
    print("Health:",health1)
    i += 1

while i == 1:
    clear_output(wait=True)
    legendName2 = input("What is the legend's name?")
    legendType2 = input("What is the legend's character type?")

    strength2 = strengthCalc()
    health2 = healthCalc()

    print("Character Name:",legendName2)
    time.sleep(1)
    print("Strength:",strength2)
    time.sleep(1)
    print("Health:",health2)
    i += 1
    
clear_output(wait=True)
print(legendName1,"Health:",health1,", Strength:",strength1)
time.sleep(1)
print(legendName2,"Health:",health2,", Strength:",strength2)
time.sleep(3)
clear_output(wait=True)
print("Let the battle begins!")

while health1 > 0 and health2 > 0:
    legend1Turn = rollDice(6)
    legend2Turn = rollDice(6)

    if legend1Turn > legend2Turn:
        damage = damageCalc(strength1, strength2)
        health2 = health2 - damage
        print(legendName1, "attacks!")
        print(legendName2, "receives",damage,"damage!" )
        print(legendName2,"has",health2,"health left!")
    elif legend1Turn < legend2Turn:
        damage = damageCalc(strength1, strength2)
        health1 = health1 - damage
        print(legendName2, "attacks!")
        print(legendName1, "receives",damage,"damage!" )
        print(legendName1,"has",health1,"health left!")
    else:
        continue

if health1 < health2:
    print(legendName2,"wins!")
else:
    print(legendName1,"wins!")


```
